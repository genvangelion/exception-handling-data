```python
import pandas as pd

df = pd.read_csv('eh_combined_eval.csv', index_col='id')
df.head()
```

```python
ex1 = df.iloc[0]
ex1
```

```python
import ast

ex1 = df.iloc[0]

exception_types = ast.literal_eval(ex1['exception_types'])
exception_types_predict = ast.literal_eval(ex1['exception_types_predict'])
print(exception_types)
print(exception_types_predict)
```

```python
# calculate for ex1

# flatten 2d lists -> get xtype vector
actual_xtype = ast.literal_eval(ex1['exception_types']) # unpack stringified 2d list -> 2d list
actual_xtype = [['a', 'b'], ['c']]
actual_xtype = [element for sublist in actual_xtype for element in sublist] # flatten 2d list -> 1d list

predicted_xtype = ast.literal_eval(ex1['exception_types_predict']) # unpack stringified 2d list -> 2d list
predicted_xtype = [['a'], ['b']]
predicted_xtype = [element for sublist in predicted_xtype for element in sublist] # flatten 2d list -> 1d list


print(actual_xtype)
print(predicted_xtype)

xtype_intersection = set(actual_xtype).intersection(set(predicted_xtype)) # cast lists as sets and get the intersection 
print(xtype_intersection)

# xtype precision = # correct / # predicted; recall = # correct / # actual
xtype_precision = len(xtype_intersection) / len(predicted_xtype)
xtype_recall = len(xtype_intersection) / len(actual_xtype)

print(xtype_precision)
print(xtype_recall)
```
